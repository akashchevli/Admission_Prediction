```python
df_copy = admission_prediction.copy()
```

```python
df_copy.head()
```

```python
df_copy[(df_copy['GRE'] >= 330) & (df_copy['TOEFL'] >= 115) & (df_copy['CGPA'] >=9.00)].sort_values(by=['Probability'], ascending=True)
```

```python
X = df_copy.loc[:,['GRE','TOEFL','University Rating','SOP','LOR','CGPA','Research']]
y = df_copy['Probability']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
import statsmodels.api as sms
```

```python
X = sms.add_constant(X_train)
estimate = sms.OLS(y_train,X)
estimate = estimate.fit()
estimate.summary()
```

```python
len(y_test)
```

```python
x = [i for i in range(1,81)]
plt.plot(x, y_test, color='green', linewidth=2, label='Test')
plt.plot(x, predictions, color='orange', linewidth=2, label='Predict')
```

```python
sns.distplot((y_test-predictions), bins=50)
```

```python
plt.scatter(y_test, predictions)
sns.regplot(y_test, predictions)
```

```python
from sklearn import metrics
```

```python
print('R Squared', metrics.r2_score(y_test, predictions))
print('MAE', metrics.mean_absolute_error(y_test, predictions))
print('MSE', metrics.mean_squared_error(y_test, predictions))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
